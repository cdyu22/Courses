```python
def isEven(integer_value):
    if integer_value%2==0:
        return True
    return False
```

```python
integer = int(input("Enter an integer: "))
if isEven(integer) == True:
    print('Yay, it is even!')
else: 
    print('No, it is odd...')
```

```python
def grade_calc(integer_value):
    if 97 <= integer_value:
        print("A+")
    elif 93 <=integer_value<97:
        print("A")
    elif 90 <=integer_value<93:
        print("A-")
    elif 87 <=integer_value<90:
        print("B+")
    elif 83 <=integer_value<87:
        print("B")
    elif 80 <=integer_value<83:
        print("B-")
    elif 77 <=integer_value<80:
        print("C+")
    elif 73 <=integer_value<77:
        print("C")
    elif 70 <=integer_value<73:
        print("C-")
    elif 67 <=integer_value<70:
        print("D+")
    elif 60 <=integer_value<67:
        print("D")
    else:
        print("F")
```

```python
grade_calc(1)
```
