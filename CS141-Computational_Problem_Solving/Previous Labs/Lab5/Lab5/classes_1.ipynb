# Object-oriented programming and classes <a id="oop"/><a id="classes"/>

Object-oriented programming (OOP) is a common approach to software development.

The idea is to create objects that encapsulate both *data* (variables), and *methods*, which are functions that act on the object's variables.

A class is a type of object that can be created and used. You are already familiar with several.

# Python is inherently object-oriented

When we talked about functions, we talked about built-in functions, imported functions, and functions we write ourselves. Classes are very much the same. Everything in Python is an object, and all objects have a type, that is, they belong to a class.

# A class you already know

Strings are an example of a type of object (a class) that is built-in to Python. Python strings are a good example of object-oriented programming.

The *data* of a string are the characters that make up the string.

The *methods* of a string are the various functions that can act on the string's characters.  [A complete list](https://docs.python.org/3/library/stdtypes.html#string-methods) of string methods can be found in the Python documentation.  Here we illustrate two string methods.

```python
sentence = 'The man in black fled across the desert, and the gunslinger followed.'

# The lower() method converts the string to lower case.
print(sentence.lower())

# The upper() method converts the string to upper case.
# This is useful for angry emails and ransom notes.
print('')
print(sentence.upper())
```

Notice the following:

```python
print(sentence)
```

Our variable <code>sentence</code> still represents the original string - it still references the initial string object which has mixed cases. When we call <code>upper()</code> or <code>lower()</code> these methods return a new string object in the appropriate case.

```python
SENTENCE_WHAT = sentence.upper()
sentence_shhh = sentence.lower()
print(SENTENCE_WHAT)
print(sentence_shhh)
print(sentence)
sentence = sentence.upper()
print(sentence)
```

Remember, a method is like a function, which means it can have a return value. The <code>upper()</code> and <code>lower()</code> methods return new objects of type str. We'll talk about this later, but strings are **immutable** objects which means that we can't change the character values that comprise them after they are created. If we want to modify the string, we have to create the modified string as a new object.

```python
print(type(sentence.lower()))
```

Notice that methods are only defined in the context of an object. The syntax for using them is:

`object_name.method_name(method_parameters)`

Contrast this with the function len:

```python
name = 'Deschain'
#Using a method
print(name.lower())
#Using a function
print(len(name))
```

# Imported classes

You can import classes from modules in the same way we've imported functions previously. Let's take a look at the Decimal class which is part of the decimal module. The documentation for this class can be found here: 
https://docs.python.org/3/library/decimal.html#decimal.Decimal

The documentation discusses some advantages to using the Decimal class over the built-in `float` type.

Using the syntax for the import statement, we can create a decimal number using the class name `Decimal`:

```python
from decimal import Decimal

my_num = Decimal('5.60')
float_num = Decimal(5.60)
your_num = Decimal(12)
print(my_num, float_num, your_num)
```

When we use the class name followed by parentheses and a set of parameters, it calls a special method in the class called the constructor. This creates a Decimal object, which is known as an `instance` of the Decimal class. That is, it is a specific object of type Decimal.

Notice that at first we create a decimal by passing in a string to the constructor:

`my_num = Decimal('5.60')`

This creates a decimal with the value and representation:

5.60

We can also create a fraction by passing in a float to the constructor:

`float_num = Decimal(5.60)`

This creates something different (remember, float representations are imprecise!):

5.5999999999999996447286321199499070644378662109375

We also used an int.
