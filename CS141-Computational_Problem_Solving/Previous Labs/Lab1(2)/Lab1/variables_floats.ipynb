# Floats <a id="floats"></a>

Floats, or, more precisely, floating point numbers, can represent any real number (e.g.  , $\pi$), not just whole numbers.  This means that they have a fractional part.

The term <b>floating point</b> has to do with the way these numbers are represented and manipulated (an important topic we will discuss in detail later).  In the familiar decimal (base-10) number system, floating point numbers like
$$
1.2345 \times 10^{2}
$$
or
$$
1.2345 \times 10^{3}.
$$
The first number represent $123.45$ while the second number represents $1234.5$.  

We can store either number using $1.2345$ and the exponent on the $10$ (either $2$ or $3$, in this case).  

In both representations the decimal point "floats" to the same location, $1.2345$.

## How is a float indicated in Python?

The presence of a decimal point '.' means that the number involved is a **float**.  

This means <code>3</code> and <code>3.0</code> are of different types.

```python
i = 3
print(type(i))
x = 3.0
print(type(x))
pi = 3.14
print(type(pi))
```

# Floating point arithmetic <a id="arithmetic"/>

Floating point arithmetic is pretty much like integer arithmetic, only the results are always floating point numbers.

## Addition 

```python
print(2.0 + 2.0)
```

## Subtraction

```python
print(54.0 - 42.0)
```

## Multiplication

```python
print(3.0 * 5.0)
```

## Division

```python
print(5.0 / 2.0)
```

## Integer division <a id="int_division"/>

This works the same way as for integers, only the result is a floating point number, even if the result is a whole number.

```python
print(8.0 // 3.0)
```

```python
print(3.14 // 0.6)
```

```python
print(9.0 // 3.0)
```

## Modulus <a id="modulus"/>

Floating point modulus can return fractional answers.  Modulus is rarely used for floating point numbers.

```python
# 3 1/8 % 1/4
print(3.125 % 0.25)
```

```python
print(3.14 % 0.25)
```

## Exponentiation <a id="exponentiation">

```python
print(4.0**2.0)
```

```python
print(49.0**(1/2))
```

### Floating point arithmetic is not always exact!

Floating point operations do not always give exact answers.  We will discuss the reasons for this later in the semester.

```python
print(0.1 + 0.2)
```

What we can guarantee is that in general, if there is an error in the answer, its magnitude is negligible compared to the magnitude of the correct answer (i.e., the **relative error** is small).  There are some exceptions to this rule, however, that one can encounter in practice.

The inexactness stems from the fact that the set of floating point numbers is finite.  This means there are gaps where there are real numbers that cannot be exactly represented by a floating point number.
