# Booleans <a id="booleans"></a>

Boolean variables are logical variables that can take on one of two values: <code>True</code> or <code>False</code>.

"Boolean" honors the 19th century English mathematician and logician [George Boole](http://www-groups.dcs.st-and.ac.uk/~history/Biographies/Boole.html).  The word is frequently not capitalized and written as "boolean".

```python
T = True
F = False
print(T, F)
```

## Boolean expressions <a id="expressions"/>

Much more interesting are <b>boolean expressions</b>.

```python
print(1 == 2)
```

The boolean expression <code>1 == 2</code> asks True or False: does 1 equal 2 (in the mathematical sense)?
<div><img style="float: left;" src="images/danger.svg" height="40" width="30">&nbsp;</div>
<div style="margin-left:0.3in;">
Make sure you know the difference between <code>=</code> and <code>==</code>:
<ul>
<li>one equal sign: assignment</li>
<li>two equal signs: mathematical equivalence</li>
</ul>
</div>

Other comparison operations are more obvious.
* <code><, <=</code> &nbsp; strictly less than, less than or equal to
* <code>>, >=</code> &nbsp; strictly greater than, greater than or equal to
* <code>!=&nbsp;&nbsp;&nbsp;</code> &nbsp; not equal to

```python
print(2 > 1)
```

```python
print(1 > 1)
```

```python
print(1 >= 1)
```

```python
A = (2 == 3)
print(A)
```

```python
x = 3
y = 4
A = x + y == 7
print(A)
```

```python
x = 3
y = 4
print(3 != 3)
print(x != y)
```

Parentheses arguably make some logical statements more intelligible, though they are not required.

```python
x = 3
y = 4
A = (x + y == 7) # breaks up all those equal signs!
print(A)
```

```python
a = 1
b = 0
print(a > b)
```

Boolean values can be stored in variables.  In the following example, the logical statement "is <code>a</code> strictly greater than <code>b</code>" is true.

```python
a = 1
b = 0
a_is_bigger = (a > b)
print(a_is_bigger)
```

Here is a more complex example of a Boolean expression, which asks whether <code>x</code> lies between -2 and 1.

```python
x = 3
x_is_in_range = (-2 <= x <= 1)
print(x_is_in_range)
```
