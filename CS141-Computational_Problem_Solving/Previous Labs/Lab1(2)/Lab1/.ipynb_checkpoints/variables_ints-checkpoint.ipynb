# Integer division <a id="division"/>

This type of division first performs standard division but then rounds down to the nearest integer.  The result of this operation is always an integer.

```python
print(8 // 3)
print(type(8 // 3))
```

Integer division is useful in situations like the following:  Suppose we have bags that will each hold 3 possums.  How many bags do we need to store 8 possums?

If we performed usual division, the answer would be <code>8/3</code>.  But this isn't right since a fractional number of bags doesn't make sense.

Instead, the correct answer is <code>8//3 + 1 = 3</code>.

Note that for negative numbers, rounding down means rounding to the next whole number to the left in the number line.  Thus,
$$
\frac{-8}{3} = -2 \frac{2}{3},
$$
which is rounded down to $-3$:

```python
print(-8 // 3)
```

<div><img style="float: left;" src="images/danger.svg" height="40" width="30">&nbsp;</div>
Observe that as a consequence, -(8 // 3) is not the same as (-8 // 3)!

```python
print('-(8 // 3):', -(8 // 3))
print('(-8 // 3):', (-8 // 3))
```

# Modulus (remainder) <a id="modulus"/>

The <b>modulus</b> operation, represented by
<pre>
    n % m
</pre>
in Python, returns the remainder when <code>n</code> is divided by <code>m</code>.

The remainder is what you would expect if the numbers involved are positive.

```python
print(8 % 3)
```

```python
print(3 % 8)
```

```python
print(0 % 8)
```

What do you think happens if we try to find the remainder after division by 0?

```python
print(8 % 0)
```

**Answer.**
<div class="voila">
<code>n % 0</code> is not defined and produces an error!
</div>

<div><img style="float: left;" src="images/danger.svg" height="40" width="30">&nbsp;</div>
If negative numbers are involved, the definition of the modulus (remainder) is a bit more complicated. Fortunately, modulus operations involving negative numbers are infrequent.
