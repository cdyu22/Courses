# Variables <a id="variables"/>

A variable is a name we use to represent an object (e.g., number, string of letters) in a program. We will talk more specifically about the relationship between objects and variables in class.

```python
message = 'Hello, world!' # message is a variable
print(message)
r = 7 # r is a variable
print('r:', r)
w=r
```

## <img style="float: left;" src="images/danger.svg" height="40" width="30">&nbsp; = means assignment <a id="assignment"></a>

The equal sign is used for assigning a value to a variable.  It does not mean equality!

The value on the right-hand side of = is associated with the variable name on the left-hand side.

This means that there is perfectly fine Python code that would make your algebra teacher gnash their teeth.

```python
x = 42
print('x is', x)
x = x + 12       # If = meant equal rather than assignment, 
print('x is', x) # this would be nonsense!
```

<p class="try_it">**Try it yourself.** Create a new variable with the value 54 and print the result.</p>

## Only variables can be assigned values.

Remember that in an assignment, the variable on the left is assigned the value on the right.

```python
x = 2 # this works
print(x)
2 = x # this doesn't
```

### Why did this error occur?

The preceding statement produces an error because the critter on the left-hand side of the assignment is what is called a **literal**.  A literal represents a fixed value.

For example, 42 is a fixed integer value, while 'Hello, world!' is a fixed string of characters.  

If we could assign values to literals, than we could change the value of 42 to be 3.14, which is probably not a good thing.

We can, however, assign a variable the value of a literal to a variable:

```python
x = 2
print(x)
```

## Variables can vary!

We can change the value of a variable as we wish.

```python
x = 3.14
print('x is', x)
x = 'Hello, world!'
print('Now x is', x)
```

## Naming rules for variables <a id="naming"></a>

<ul>
<li>
    Variables can only contain letters, numbers, and underscores.
    <ul>
    <li><code>x, x2, hello_world</code> are all ok</li>
    </ul>
</li>
<li>
    Variable names can start with a letter or an underscore, but not with a number.
    <ul>
    <li><code>_retcd</code> is ok; <code>2x</code> is not</li>
    </ul>
</li>
<li>
    Spaces are not allowed in variable names.  Use underscores instead.
    <ul>
    <li>e.g., <code>hello_world</code> is ok, but <code>hello world</code> is not</li>
    </ul>
</li>
<li>
    Variable names should be descriptive, though not ridiculously long.
</li>
</ul>

```python
# Note that in the print statements we are using = in its 
# mathematical sense, rather than as representing assignment!

hello_world = 3.14
print('hello_world =', hello_world) 
x2 = 'hello, world'
print('x2 =' , x2)
_retcd = 1
print('_retcd =', _retcd)
__ = 42
print('__ =', __)
```

## Keywords: some words are off-limits as variable names

In every computer language there are certain words called <b>keywords</b> that cannot be used for variables because they are already used as parts of the language. Here are the 
<a href="https://docs.python.org/3/reference/lexical_analysis.html#keywords">Python keywords</a>:

<pre>
False    class      finally   is         return
None     continue   for       lambda     try
True     def        from      nonlocal   while
and      del        global    not        with
as       elif       if        or         yield
assert   else       import    pass
break    except     in        raise
</pre>

Let's see what happens if we try to use keywords as names of variables.

```python
True = 42
```

```python
return = 3.14
```
