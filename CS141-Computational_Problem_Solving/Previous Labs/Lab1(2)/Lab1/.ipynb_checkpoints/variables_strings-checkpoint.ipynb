 # Strings <a id="strings"></a>
 
Strings, or character strings, are just that &ndash; strings of characters.  Python refers to the string type as <code class="kw">str</code>.

# Single and double quotes

Strings are indicated by either single or double quotes.  The fact that we can use either allows us to place quotations inside strings.

```python
x = 'Hello, world!'
print(x)
y = "Hello, world!"
print(y)
z = "With a last gasp he said, 'It was the squirrels!'and 'aha!'"
print(z)
```

You can either delimit a string with single quotes or double quotes, but you can't use different types of quotes to delimit a character string.  

Suppose you could: what would have happened in the assignment to **z** we just did?

**Answer.**
<div class="voila">
The right-hand side would consist of the string
<pre>
'With a last gasp he said, "
</pre>
followed by 
<pre>
It was the squirrels!"'
</pre>
However, the latter is not a string since it lacks a quotation mark at the beginning.  This would cause an error.
</div>

Python will let you know if you fail to correctly delimit a string as soon as it recognizes your mistake.  In the following example it has to go to the end of the line before it can be sure the string is not <code>"foo''"</code>.

```python
# This will produce an error.  
# "EOL" stands for "end of line".
x = "foo'
```

# Combining strings (concatenation) <a id="concatenation"></a>

We can combine strings using the **+** operator.

The term typically used in CS for combining strings is **concatenation**, which the [Oxford English Dictionary](http://www.oxforddictionaries.com/us/definition/american_english/concatenation)
defines as "the action of linking things together in a series".

```python
hello = "hello"
world = "world"
space = " "
hello_world = hello + space + world
print(hello_world)
a=3+7
b=3
c=7
d=b+c
print (a,d)
```

# String multiplication <a id="multiplication"><a>

You can obtain repetitions of a string by multiplying it by a whole number.

```python
s = 'buffalo \t'
print(3 * s)

n = 8
print(n * s)

t = 4 * s
print(t)

print(9 * '$')
print(9 * '$,')
```

# Whitespace and the escape character \ <a id="whitespace"></a>

The term <b>whitespace</b> refers to characters that are legitimate characters, but which are invisible. 

The most common whitespace characters are spaces, tabs, and newlines.

Spaces are represented in the obvious way using spaces, <code>' '</code> or <code>'  '</code>.

Tabs are represented by one or more <code>'\t'</code>.

Newlines, which indicate the end of lines and cause a line break to appear in output, are represented by one or more <code>'\n'</code>.

In this context the backslash in the representation of a tab or newline is called an <b>escape character</b>.  An escape character changes the intepretation of the character that follows.

```python
print("This string has a triple space before the last   word.")
```

```python
print("\tThis string begins with a tab.")
```

```python
print("This string contains \n a newline that causes a line break.")
```

```python
print('manbear\npig')
```

```python
print(9 * '$\n')
```

# Printing a backslash <a id="backslash"></a>

Since the backslash is interpreted as an escape character, in order to print the backslash character you need to use <code>\\\\</code> (i.e., you apply the escape character to the backslash).

```python
print('This is how to print a backslash: \\n')
```

# Getting input from the user <a id="input"></a>

Since there is no telling what users might type when prompted, Python's **input** function reads everything you type as a string.

```python
s = input('What is your name? ')
print("Hi " + s)
```

Observe that we leave a space at the end of the prompt.  If we don't, our response is munged together with the prompt and the result is ugly:

```python
s = input('What is your name?')
t = input('What is your quest?')
print("Good luck with",t,s)
```

```python
s = input('What is the airspeed of a European swallow? ')
t = input('What is the airspeed of an African swallow? ')
print('Combined airspeed:', s + t)
```
