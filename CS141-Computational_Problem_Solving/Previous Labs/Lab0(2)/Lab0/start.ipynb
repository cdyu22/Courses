# About Python

The Python programming language dates to the late 1980s.  It was originally developed by Guido van Rossum, then at the Centrum Wiskunde &amp; Informatica in the Netherlands.

The popularity of Python in big data and data science has placed Python first on the Institute of Electronic and Electrical Engineers (IEEE) ([2017 list of top programming languages](http://spectrum.ieee.org/computing/software/the-2016-top-programming-languages)).

One of the great strengths of Python is its enormous software ecosystem which makes it possible to do interesting things without having to write lots of code.  You can see a few examples [here](cool_stuff.ipynb).
