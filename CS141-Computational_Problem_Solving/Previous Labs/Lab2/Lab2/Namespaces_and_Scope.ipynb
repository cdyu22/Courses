<h1 style="text-align:center;">A few notes on scope and namespaces</h1>

Python uses the *LEGB* rule to search namespaces to located variable name to object mappings. This stands for:

* Local
* Enclosed
* Global
* Built-in

Python searches namespaces in these scopes in this order.

*Local* for our purposes at the moment refers to inside a function.

*Enclosed* refers to situations where, for example, a function is enclosed inside another function.

*Global* refers to the main namespace.

*Built-in* refers to a set of names that Python has reserved for its own use. 

Consider the following:

```python
def foo():
    my_var = 'In the function'
    print('my_var is',my_var)
    
    
    
my_var = 'In the main program'
foo()
print('my_var is',my_var)
```

The variable <code> my_var </code> is defined both locally in the function and globally in the main program. When we are in the function, Python first checks the local namespace to find the value of my_var, and it prints the local value. Outside of the function, we get the global value.

What happens if the variable is only defined globally?

```python
def foo():
    print('my_var is',my_var)



my_var = 'Globally defined'
foo()

print('my_var is',my_var)
```

Notice that when <code> my_var </code> is only defined globally, the print statement in the function prints the global value. Python first checks the local namespace, and when it finds no entry, it checks the global namespace (the function is not enclosed so we are skipping E).

When you make an assignment, as opposed to just printing out the value, Python assigns it to the local scope and this "overshadows" the global assignment, leading to this UnboundLocalError.
```python
def foo():
    my_var = 'Outer function' #What happens if we comment this out?
    print("my_var in foo()", my_var)
    def bar():
        my_var = 'Inner function' #What happens if we comment this out?
        print("my_var in bar()", my_var)
    bar()

my_var = 'Main program'
foo()
print("my_var in main program", my_var)
#bar() 
```

The function bar() is enclosed in the function foo(), so Python will first check the local namespace of bar(), if it doesn't find anything, it will check the enclosing namespace - that of foo(), and then if it still doesn't find anything it will check the global namespace.
