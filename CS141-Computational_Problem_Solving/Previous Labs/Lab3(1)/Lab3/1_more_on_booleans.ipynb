# Logical operations

The basic operations on boolean variables are 
* <code>not</code>, 
* <code>and</code>,
* <code>or</code>.

# Logical NOT <a id="not"/>

If <code>A</code> is a boolean variable, then <pre>not A</pre> is <code>True</code> if  <code>A</code> is false, and <code>False</code> if <code>A</code> is true.


```python
print("not True: ", not True)
print("not False:", not False)
```

```python
smelly = True
print(not smelly)
```

```python
slimy = False
print(not slimy)
```

```python
stmt = (2 > 1) # 2 > 1 is true.
print(stmt)
print(not stmt)
```

# Logical AND  <a id="and"></a>

If <code>A</code> and <code>B</code> are two boolean variables, then <pre>A and B</pre> is <code>True</code> if **both** <code>A</code> and <code>B</code> are true, and <code>False</code> otherwise.

```python
print("True and True:  ", True and True)
print("True and False: ", True and False)
print("False and True: ", False and True)
print("False and False:", False and False)
```

```python
print((2 > 1) and (1 > 1)) # (2 > 1) is true but (1 > 1) is false.
```

```python
print((2 > 1) and (1 >= 0)) # (2 > 1) is true and (1 >= 0) is true.
```

```python
x = 3
y = 4
A = (x + y > 6) # true, since x + y = 7
B = x*y > 21    # false, since x*y = 12
print(A and B)
```

# Logical OR <a id="or"/>

If <code>A</code> and <code>B</code> are two boolean variables, then <pre>A or B</pre> is <code>True</code> if **at least one** of <code>A</code> and <code>B</code> is true, and <code>False</code> otherwise.

Thus,
<pre>
(it is raining) or (it is not raining)
</pre>
is **always true** since one of the two statements must be true.

This is not the everyday meaning of "or".  In everyday speech when we say "or", we imply a choice of two statements, **only one of which is true** (this is called the <b>exclusive OR</b>, or <b>XOR</b>).

Thus, if you are a true computer scientist and someones asks a question like, "Do you want pizza tonight, or not?", you should always answer "Yes!", since one of those two statements must be true.  This is a very good way of annoying friends and loved ones and it never gets old.

```python
print("True or True:  ", True or True)
print("True or False: ", True or False)
print("False or True: ", False or True)
print("False or False:", False or False)
```

```python
T = True
F = False
print(T or F)
```

```python
print(T or T)
```

```python
print(F or F)
```

```python
print((2 > 1) or (1 > 1)) # (2 > 1) is true but (1 > 1) is false.
```

```python
print((2 > 1) or (1 >= 0)) # (2 > 1) is true and (1 >= 0) is true.
```
