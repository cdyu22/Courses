```python
class w_class(object):     
    def hide(self):         
        print('Burying my head in the sand.')    
class t_class(w_class):     
    def move(self):         
        print('Follow the moon!')
class c_class(t_class):     
    def hide(self):         
        print('Going inside my shell.')     
class q_class(c_class):     
    def move(self):         
        print ('Slow and steady wins the race!') 

class a_class(c_class):     
    def move(self):         
        print('Time to make the donuts!') 


```

```python
one = q_class() 
two = c_class() 
three = w_class() 
four = a_class() 
five = t_class() 

one.move() 
two.move() 
three.move()
```
