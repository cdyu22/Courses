```python
def func_one(a):
    result = 0
    while (a!=0):
        result += a % 10
        a = a // 10
    return result
```

```python
func_one(153215)
```

```python
def play_nims(pile, max_marbles):
    print("The number of marbles in the pile to start is: " + str(pile))
    print("The maximum number of marbles the players can take on one turn is: " + str(max_marbles))
    whose_turn = 0
    while (pile > 0):
        print("The player " + str(whose_turn) + "'s turn:")
        marbles = int(input("How many marbles do you want to take?"))
        while (marbles <= 0) or (marbles > max_marbles) or (marbles > pile):
            marbles = int(input("Invalid input! How many marbles do you want to take?"))
        pile -= marbles
        print("The pile has " + str(pile) + " marbles in it.")
        if (pile == 0):
            print("The player " + str(whose_turn) + ' wins!')
            pile -= max_marbles
        whose_turn = (whose_turn + 1) % 2
```

```python
play_nims(20,5)
```

```python
s = 'I had a cat named amanda when I was little'
count = 0
for c in s:
    if (c == "a"):
        count += 1
print(count)
```
