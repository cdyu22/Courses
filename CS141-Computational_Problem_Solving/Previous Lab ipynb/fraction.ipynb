```python
class Mage:
    def __init__(self,name):
        self.name = name
        self.ability_level = 1
        self.power = 100
    
    def cast(self, spell_points):
        if(self.power > spell_points):
            return True
        return False
    
    def get_power(self):
        return self.power
    
    def increase_ability(self):
        self.ability_level += 1
wizard = Mage('Sir Castalot')
```
