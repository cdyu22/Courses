```python
class Battery():
    
    def __init__(self, capacity):
        self.__max_charge = capacity
        self.__charge_remaining = capacity
        
    def get_max_charge(self):
        return self.__max_charge
    
    def get_charge_remaining(self):
        return self.__charge_remaining
    
    def get_percent_charge(self):
        self.__percent_charge = ((self.__charge_remaining)/(self.__max_charge))*100
        return self.__percent_charge
    
    def discharge(self, charge):
        if charge > self.__charge_remaining:
            self.__charge_remaining =  0
            print('Greater than charge remaining, battery fully discharged')
        else: 
            self.__charge_remaining -= (charge)
    
    def recharge(self, charge):
        if charge > self.__max_charge - self.__charge_remaining:
            self.__charge_remaining = self.__max_charge
            print('Greater than maximum charge, battery fully charged')
        else:
            self.__charge_remaining += (charge)

    def __str__(self):
        return "Battery with maximum capacity: " + str(self.__max_charge) + ' and ' +str(self.get_percent_charge())+' percent charge remaining.'
```

```python
energizer = Battery(400)
#Energizer is an instance
```

```python
energizer.discharge(150)
```

```python
energizer.get_charge_remaining()
```

```python
energizer.discharge(300)
```

```python
energizer.recharge(250)
```

```python
print(energizer)
```

```python
#Accessor Method: We're not trying to change instance variables.
#Mutator: Inside of the method definition we're trying to modify the instance variables. 
#Accessor in Battery: max_charge, charge_remaining, get_percen_charge, __str__
#Mutator in Battery: Discharge and Recharge
```
