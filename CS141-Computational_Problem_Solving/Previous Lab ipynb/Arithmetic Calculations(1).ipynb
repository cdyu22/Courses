### A few notes on notebooks

We'll talk about this in the lab in more detail, but a few things to keep in mind:

-To run a code cell, put the cursor in the cell and press shift-Enter or the Play button in the tool bar
-Closing the tab a notebook is running on just takes it out of your view - it doesn't stop the process running the notebook. To do this, you should go to File---> Close and Halt. This is important as too many open files will cause your notebook server to crash.
-If you want to clear the output and start over, take a look at the Kernel menu. There's a few options there including Interrupt, which is useful if you accidentally start an infinite loop!

```python
# Now let's try to compute the area of a circle with radius 10
3.14159 * 10 ** 2
```

We could write an algorithm for computing the area of a circle as:

1) Square the radius

2) Multiply the result from step 1) by the mathematical quantity pi

The input is the radius and the output is the area. The numbered steps comprise the algorithm.

```python
import math
r = 10
area = (r**2)*math.pi #r**2 means r to the second power and math.pi is the value 3.1415927....
print(area)
```
