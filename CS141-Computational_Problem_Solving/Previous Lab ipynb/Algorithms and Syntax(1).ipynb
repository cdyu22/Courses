We could write an algorithm for computing the area of a circle as:

1) Square the radius

2) Multiply the result from step 1) by the mathematical quantity pi

The input is the radius and the output is the area. The numbered steps comprise the algorithm.

As you are writing and debugging code, we will often discuss algorithmic versus syntactical (i.e. dealing with syntax) errors. Syntax, by definition, is the set of rules that govern what comprises correctly structured, understandable, and executable code. Syntax errors will generally result in the program crashing because the Python interpreter cannot understand what you mean.

An algorithm (among other things) is a process or a set of rules to be followed in calculations or other problem-solving operations. Algorithmic errors are often present in programs that execute successfully. 

Here is a correct implementation in Python of the area algorithm for a radius defined by the variable r:

```python
import math
r = 10
area = (r**2)*math.pi #r**2 means r to the second power and math.pi is the value 3.1415927....
print(area)
```

There are two more implementations below. One has an algorithmic error and the other has a syntactical error. Which is which?

```python
#Program A
import math
r = 10
area = (r**2)+math.pi 
print(area)
```

```python
#Program B
import math
r = 10
area = (r**2)math.pi
print(area)
```
