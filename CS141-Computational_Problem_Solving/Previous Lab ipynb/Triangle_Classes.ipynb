```python
class Triangle(object):
    
    def __init__(self, base, height):
        self.base = base
        self.height = height
    
    def area(self):
        return 0.5 * self.base * self.height
    
    def represent(self):
        print('This is a triangle with a base like this:', self.base * '*')
        print('It is this high: ')
        for i in range(self.height):
            print('*')
```

```python
tri = Triangle(2,10)
```

```python
tri.represent()
```

```python
class Right_Triangle(Triangle):
    def get_hypotenuse(self):
        return (self.base**2 + self.height**2)**.5
```

```python
rtri = Right_Triangle(2,10)
print(rtri.area())
print(rtri.get_hypotenuse())
```

```python
class Equilateral_Triangle(Triangle):
    def __init__(self,side):
        #self.base = side #Overloading, making the self.base side. We can also use a superfunction. Which goes back to the parent class
        #And if we want to call any method in the class we can use the super function and the method name in the class.
        #self.height = (side**2-(side/2)**2)**.5
        super().__init__(side,(side**2-(side/2)**2)**.5)#Instead of redefining them one by one we can use a single line to redefine all of the instance variables.
    def represent(self): #we want to rewrite the old method, so now we create a new method of the same name. Has to be same
        for i in range(1, self.base+1):
            print(' '*(self.base-i) + '* ' * i)
```

```python
#can access all methods from superclass, but can't access childclass methods from superclass.
```

```python
equi = Equilateral_Triangle(12)
print(equi.area())
(equi.represent())
```

```python
class Isoceles_Right(Right_Triangle):
    def __init__(self, side):
        #self.base = side
        #self.height = side
        #or
        super().__init__(side,side)
    def represent(self):
        for i in range(1, self.height+1):
            print(i*'* ')
```

```python
iso_ri = Isoceles_Right(5)#Inheritance is from all parent and grandparent classes.
```

```python
print(iso_ri.get_hypotenuse())
print(iso_ri.area())
iso_ri.represent()
```
