```python
class Four(object):

    def __init__(self, x = 3, y = 7):
        self.__one = x
        self.__two = y

    def setOne(self, x):
        self.__one = x

    def setTwo(self, y):
        self.__two = y

    def getOne(self):
        return self.__one

    def getTwo(self):
        return self.__two

    def __str__(self):
        return str(self.__one) + '*' + str(self.__two) + '=' + str(self.__one * self.__two)
instance1 = Four(5, 6)
instance2 = Four(4)
```

```python
instance1.setTwo(9)
```

```python
print(instance1.getTwo() - instance2.getTwo())
```
