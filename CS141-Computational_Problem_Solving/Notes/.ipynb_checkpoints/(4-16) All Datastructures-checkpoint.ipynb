```python
#Symbols, Muttable, How to access the elements?
```

```python
#String

#Symbol: " " or ''' '''

#Muttable: No
#Example, can't run
a = 'hello'
a[0] = 'H'
#Note that a.upper() doesn't change the string, it just makes it upper in that instance.

#How to access: Index
```

```python
#List

#Symbol: [ ]

#Muttable: Yes

#How to access: Index, so integers starting from 0
```

```python
#Tuple

#Symbol: ( )

#Muttable: No

#How to access: Index, so the  same as the string
```

```python
#Dictionary

#Symbol: { } (Same as set, but uses :) {Key:Value}

#Muttable: Yes. Can run:
aa = {1:34}
aa[1] = 100
aa

#How to access: With the key, there's no integer starting from 0
```

```python
#Set

#Symbol: { }

#Muttable: Yes
#Need to call some boolean method to add and call things in sets. So
s = {1,2,3}
#s[0] = 100 Doesn't work
s.add(100)
s.remove(1)
s

#How to access: None
```

```python
#Object for chess piece, want to create a chessboard and populate it with several instances of these objects. 
#Some squares on the board will be empty, containing objects of type None. Which data structure works best?

#You need it to be mutable. Need it to be easy to access any locations inside of te grid.

#Dictionary would work, as it's muttable.Can just give each square a key value when defining the entire board. Square is is
#Some value. Can define all of the values using a unique key. 
#Limitation is if it's a small board, 3x3 or 4x4. But if it's 100x100 and you want to define all the squares need to define
#ALL of the squares, takes a lot of time.

#Tuples doesn't work, as it's not muttable.

#Lists could theoretically work, if it's fairly small. Could associate each square with integers, but if it's large, 100x100,
#Have to go through each one.

#Set is muttable, but not associated with anything, so doesn't work.

#List of lists work, its basically two-dimensional. 
#this is actually the best choice, as having a row and column value makes it easier to access any given place.
```

```python
#You grade student exams based on student ID numbers, want to retrieve student's name by ID number after grading the exams. 
#Which object works best?

#Data structure should be muttable, as no exams yet, so score should be 0, then need to change it later.
#Want to retrieve student name by ID number. 

#Dictionary. 
#But can any data structures work.

#Tuple won't as it's immutable, and we need to update the score later.

#Lists could theoretically work, as lists are muttable, can use the index value's position, but not always in practice.
#As WM ids go up to 11 digits and any that aren't occupied by students would be wated, it's not space efficient.

#Set wouldn't work, as there's no sequence.

#List of Lists is two dimensional, could theoretically work, but would need to make a really large list of lists to hold value
#in just a small part, so not space efficient.
```
