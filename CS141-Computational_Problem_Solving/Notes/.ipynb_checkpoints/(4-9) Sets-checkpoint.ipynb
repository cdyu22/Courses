```python
#An unordered collection with no duplicate elements.
#Meaning that if we have a set defined as:
#{1,2,3} = {3,2,1} = {1,3,2}
#Duplication doesn't count. So whenever we see a set containing duplicate elements, we treat it as one element.
#So {1,1,2,3} = {1,2,3}
```

```python
prime = {2,3,5,7,11,13,17,19}#dictionaries are pairs, sets are all equal to each other.
#The way we define it each other, curly brackets in pairs means dictionaries. 
#In sets we only need to cast elements seperated by commas.
print(type(prime))
```

```python
aa = {}# By default, this is a dictionary
print(type(aa))
bb = set()#This, by default, leaves it a set.
print(type(bb))
ll = []
print(type(ll))
ll = list()
print(type(ll))
ll = list('hello')
print(ll)
aa = set('hello')#As ll repeats, only one in set
print(aa)
```
