```python
import turtle
class Equilateral_Triangle():
    def __init__(self,side):#this is a function, but we call it a method. Self isn't an argument, but it's required(?).
        #This has one argument
        self.__side_length = side #Instance variable, essential for equilateral triangle.
        self.__height = (side**2 - (side/2)**2)**.5#The two underscores lets python know they're private names,
        #can's access them directly outside of the class.
    def get_side(self):#Must always use self(?)
        #Wants to tell the user the value of the side.
        return self.__side_length#Instance variables that are __ are accessable anywhere inside the class.
    #Side variables are local variables, only accessable inside the method you need.
    def get_height(self):
        return self.__height
    #Can provide more operations
    def area(self):
        return (self.__side_length)*(self.__height)*.5
    def turtle_drawing(self):
        bob = turtle.Turtle()
        for i in range(3):
            bob.forward(self.__side_length)
            bob.right(120)
    def __str__(self):#How you want to print the instance on the screen. 
        #For example, this equilateral triangle returns the string method.
        #Only need to return a string, but it will describe side and height.
        return "The side of the triangle is " + str(self.__side_length) + " and the height of the triangle is " + str(self.__height) + "."
```

```python
equi_tri = Equilateral_Triangle(100)#Put in value inside equilateral_triangle, want an instance
```

```python
print(type(equi_tri))#Defind as equilateral triangle object. When you use assignment statement it'll look for the name in its
#namespace. It'll see it's a classname and look for an init method. Will then automotically exclude the method. 
#Need to pass the value to match
```
