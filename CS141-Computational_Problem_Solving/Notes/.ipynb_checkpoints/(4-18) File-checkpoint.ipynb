```python
import random
with open('numbers_2.txt','w') as f:#After as is the object's name.It's the same as f=open('numbers.txt','w')
    #will open a new object with f.
    for i in range(100000):
        score = random.randint(0,100)
        #f.write(str(score)+'n')#instead of this, we can use
        print(score,file = f)
```

```python
#Numbers.txt
#Find min, max, and sum
#Calculate frequencies of each value
#Find value with the largest frequency
import random
f = open('numbers.txt','w')
a = []
for i in range(100000):
    score = random.randint(0,100)
    f.write(str(score) + '\n')
    a.append(score)
print(max(a))
print(min(a))
print(sum(a))
f.close
```

```python
data = open('numbers3.txt','r')
total = 0
maximum = 100
minimum = 0
score = data.readline()
while score != '':
    score = int(score)
    total += score
    if minimum > score:
        minimum = score
    if maximum < score:
        maximum = score
    score = data.readline
print('The Sum is' , total)
print('The Maximum is, ' + str(maximum))
print('The Minimum is, ' + str(minimum))
```
