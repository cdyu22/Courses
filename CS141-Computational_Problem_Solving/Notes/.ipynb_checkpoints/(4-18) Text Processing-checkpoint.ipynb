```python
#If you have a large set of data stored in another file you need to access that file from python and retrieve the data from
#there, then process it from there. So there's a building function we can use in python to open a file called open.
#One required thing
open('/Desktop/Computational_Problem_Solving/scores_1.txt')
#Python can then access that txt file by searching for it. It's returning value is a file object, meaning python will return
#it as a file object. So can do:
f = open('/Desktop/Computational_Problem_Solving/scores_1.txt')

f.readline()#will return '85/n'
#This can be used to recognize a line break. Whenever you have several lines after a current line.
```

```python
print(f.readline())#will sequence throughout the list. So it will go from the first item to the next item.
#After all of the lines have been run though, it will return an empty string, ''
#which is very useful, as if it returns an empty string we've ran through the entire file.
```

```python
#mode for opening file. We need to tell which mode we're opening the file in, by default it's read. so 'for read' 
#mode means you can't modify anyhthing in the original file, you can only access it. 'r' means read only mode.
f = open('scores_1.txt', 'r')#So it's in read only mode. If you want to change the contents in the file, you need to open it for
#for write mode, so use 'w'.
```

```python
score = f.readline()#initially, opening the file the cursors at the top of the file.
#After this, can use a while loop to check if a score is equal to an empty screen.
ll = list()
score = f.readline()
while score != '':#when it's empty it means that the file is done.
    score = str(score)
    ll.append(score)
    score = f.readline()
print(ll)
total = 0
for item in ll:
    total += item
print(total/len(ll))
```

```python
total_value = 0
counter = 0
score = f.readline()
while score != '':
    if score != '\n':
        score = int(score)
        total_value += score
        counter += 1
        score = f.readline()
print(total_value)
```
