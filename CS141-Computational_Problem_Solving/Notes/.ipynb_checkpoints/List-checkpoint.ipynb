```python
def mean(list_1):
    total = 0
    for i in list_1:
        total = total + i
    return total/len(list_1)
```

```python
my_list = [10,20,30]
print(mean(my_list))
```

```python
def median(data):
    data.sort()
    if len(data) % 2 == 0:
        return (data[len(data)//2] + data[len(data)//2-1])/2
    else:
        return data[(len(data)-1)//2]
```

```python
list_1 = [2,10,23,4,0,78]
print(median(list_1))
```

```python
class Grades(object):
    
    def __init__(self,name):
        self.__name = name
        self.__scores = []
        
    def get_name(self):
        return self.__name
    
    def add_score(self, score):
        #self.__score += [score]
        self.__scores.append(score)
    
    def get_scores(self):
        return self.__scores
    
    def get_max_score(self):
        self.__scores.sort()
        return sorted(self.__scores)[-1]
```

```python
student = Grades('Alice')
print(student.get_scores())
```

```python
student.add_score(80)
student.add_score(100)
student.add_score(90)
```

```python
(student.get_scores())
```

```python
student.get_name()
```

```python
print(student.get_max_score())
print(student.get_scores())
```
