```python
class Triangle(object):
    
    def __init__(self, base, height):
        self.base = base
        self.height = height
    
    def area(self):
        return 0.5 * self.base * self.height
    
    def represent(self):
        print('This is a triangle with a base like this:', self.base * '*')
        print('It is this high: ')
        for i in range(self.height):
            print('*')
```
