```python
data = open('numbers.txt','r')
total = 0
minimum = 100
maximum = 0

score = data.readline()
while score != '':
    score = int(score)
    total += score
    if minimum > score:
        minimum = score
    if maximum < score:
        maximum = score
    score = data.readline()

print('The sum is:' + str(total))
print('The maximum is:' + str(maximum))
print('The minimum is:' + str(minimum))
```

```python
data = open('numbers.txt','r')
freqs = [0 for i in range(101)]
score = data.readline()
while score != '':
    score = int(score)
    freqs[score] += 1
    score = data.readline()
print('The frequency is:', freqs)
```

```python
max_freq = 0
counter = -1
for i in range(len(freqs)):
    if freqs[i] > max_freq:
        max_freq = freqs[i]
        counter = i
print('The number with the largest freqs is: ' + str(counter))

        
```

```python
data = open('Bovary_Excerpt.txt','r')
ll = list()
line = data.readline()
while line != '':
    ll.append(line)
    line = data.readline()
ll = str(ll)
ll = ll.lower().replace(',','').replace('.','').replace('-','').replace('"','').replace('\n','')
```

```python
file = open('Bovary_Excerpt.txt','r')
word_counts = {}
for line in file:
    tokens = line.lower().replace(';',' ').replace(',',' ').replace('!',' ').replace('.',' ').replace('(',' ').replace(')',' ').replace('-',' ').replace('?',' ').replace('\'','').split()
    for item in tokens:
        try:
            word_counts[item] += 1
        except:
            word_counts[item] = 1
print(word_counts)
```

```python
word_lists = {}
for word,count in word_counts.items():
    try:
        word_lists[count] += [word]
    except:
        word_lists[count] = [word]
print(word_lists)
```
