```python
#Tuples are like list, only difference is they're enclosed with parentheses, the first difference.
#Tuples are data structures that are immutable.If you have a list object in python, let's say ll.
#ll = [traingle, square, face, star], if you want to replace star with a circle, you can use an index value to find the location
#Star is in fourth location, so ll[3] 
#ll[3] = circle
#Will change the list, meaning the list is mutable, mutator.
#Tuples aren't mutable. Meaning, if you don't want someone to be able to change values, you make it a tuple.
#1) ()
#2) Immutable.

#shapes = (triangle, square, face, star)
#If you try to access the list it works the same as a list, you can still access the symbols through
#shapes[3] however, if you want to change star to circle, you'll get an error, as you can't change it, it's immutable.
```

```python
point_1 = (1,2)
print(point_1[1])
```

```python
point_1[0] = 100
```

```python
funlist = [(1,2,3),(4,5,6)] #A list object.
print(funlist[1])
```

```python
print(funlist[1][1])
```

```python
funlist[1] = (7,8,9)
print(funlist)
```

```python
#Dictionary
#Dictionaries hold the values and map them to some key, but instead of using integers as an index value, you need to explicitly
#set it's index value.
#Curly Bracket
my_dictionary = {10:'A',15:'B',8:'C'}#can be any other data types, such as strings or variable names.
#key value is 10, A is the value associated with the key 10.
print(my_dictionary[10])
```

```python
print(my_dictionary[0])
```

```python
my_color = {'Alice':'Green', 'Rebecca':'Teal'}
print(my_color['Alice'])
print(my_color['Bob'])
```

```python
my_color['Conner'] = 'Orange'#can add more key value pairs after it's set.
#Can you add any to tuples?
```

```python
print(my_color)
```
