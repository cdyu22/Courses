```python
#Loop One
for i in range(10):
    print(i)
print('Done')
```

```python
#Recreate Above with while loop
i=0
while i < 10:
    print(i)
    i += 1
print('Done')
#Must initialize i before while loop. While loop is more flexible. But for loop 
#is predictable. If you know a sequence, use for loop, otherwise, use while loop.
```

```python
#Loop Two
i=0
while i < 10:
    if i % 2 == 0:
        i *= 2
    print(i)
    i = i+1
print('Done')
```

```python
#Loop Three
for j in range(0,10):
    if j % 2 == 0:
        j=j*2
    print(j)
print("Done")
```

```python
#Loop Four
for i in range(5):
    print('i = ',i)
    for j in range(3):
        print('j = ',j)#Finishes inner layer first, then repeats it until outer loop is done.
```

```python
#Loop Five
for i in range(5):
    print('i = ',i)
    for j in range(i):
        print('j = ',j)
```
