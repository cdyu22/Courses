```python
#One third of the product of a,b,c
def one_third(a,b,c):
    D = a*b*c
    D = D/3
    return D
#or
def product(a,b,c):
    product =  a*b*c/3
    return product
```

```python
one_third(3,3,3)
```

```python
#2 x quotieny of (n/d)
def two_times_quotient(n,d):
    A = (n/d)
    A = A*2
    return A
#or
def double_quotient(n,d):
    return ((n/d)*2)
```

```python
two_times_quotient(10,5)
```

```python
#square of sum of e,f
def square_of_sum(e,f):
    A = e+f
    A = A**2
    return A
#or
def square_sum(e,f):
    return ((e+f)**2)
```

```python
square_of_sum(4,5)
```
