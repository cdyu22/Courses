```python
#An unordered collection with no duplicate elements.
#Meaning that if we have a set defined as:
#{1,2,3} = {3,2,1} = {1,3,2}
#Duplication doesn't count. So whenever we see a set containing duplicate elements, we treat it as one element.
#So {1,1,2,3} = {1,2,3}
```

```python
#Set: An unordered collection with no duplicate elements.
#{1,1,2,3} = {1,2,3} Count the distinct elements one by one. 
#No duplicate elements.
#Use {} with numbers.
primes = {2,3,5,7,11,13,17,19}
print(type(primes))
```

```python
A = {1,2,3}#A is a subset of B if all of the elements of A are in B
B = {1,2,3,4,5,6}#If it is, we then record it as A ⊆ B (A is the subset of B)
A <= B #If A is a subset of B, it will return true, if it isn't it'll be false
ACB#A is the proper subset of B
#A is proper subset if it's a subset and B also has additional elements.
#A can also be equal to B.
A<B#If it's a proper subset it'll return True, otherwise false.
A==B#Returns true if equal, otherwise false.
```

```python
A = {1,2,3} #A is a proper subset of B
B = {1,2,3,4,5,6}
C = {2,3,6}#Also satisfies proper subset.
print(A<B)
print(B<C)#B is the superset of C.
```

```python
#Union operation. Union of A and B will return the elements of A and B
#AUB (U is union) will give new set giving new elements of A and B
#intersection will give elements that are also A and B
#Difference Operation A-B will give the difference of A and B.
#Will return b 
#B-A will be a set containing A 
```

```python
#Union
A = {1,2,3}
B = {1,2,3,4,5,6}
print(A|B)
print(A.union(B))
```

```python
#Intersection
print(A&C)
print(A.intersection(C))
```

```python
#Difference
print(B-C)
print(B.difference(C))
```
